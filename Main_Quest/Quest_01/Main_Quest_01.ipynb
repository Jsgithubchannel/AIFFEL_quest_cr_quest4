### [메인 퀘스트 1번: 은행계좌 만들기 - 3점]

```python
import random

class Account():

  account_count = 0
  account_list = []

  def __init__(self, depositor, balance):
    self.bank_name = "SC은행"
    self.depositor = depositor
    self.account_num = self.generate_account_num()
    self.balance = balance

    # 계좌 개설(객체 생성)시 카운트 1 추가
    Account.account_count += 1

    # 객체 생성시 리스트에 정보 저장
    Account.account_list.append(self)

    print(f"\n *** 안녕하세요. {self.depositor}님의 계좌 정보 입니다. ***")
    print(self)

    # 입출금 내역 저장 리스트
    self.deposit_list = []
    self.withdraw_list = []

  def __str__(self):
    return f"은행 이름: {self.bank_name}, 예금주: {self.depositor}, 계좌번호: {self.account_num}, 잔고: {self.balance:,}원"

  # 은행 계좌 000-00-000000 랜덤 생성
  def generate_account_num(self):
    first_part = random.randint(100, 999)
    second_part = random.randint(10, 99)
    third_part = random.randint(100000, 999999)
    return f"{first_part}-{second_part}-{third_part}"

  # 개설된 계좌 조회
  @classmethod
  def get_account_num(cls):
    return f"\n총 개설된 계좌의 수는: {cls.account_count} 개 입니다."

  # SC은행 전체 계좌 조회
  @classmethod
  def account_display(cls):
    print("\n*** SC은행 전체 계좌 list ***")
    for i in cls.account_list:
      print(i)

  # 입금
  def deposit(self, money):
    if money >= 1:
      print("\n*** SC은행 입금 현황 ***")
      self.balance += money
      print(f"{money}원 입금 하셨습니다. \n현재 {self.depositor}님의 잔고는 {self.balance:,}원 입니다.")

      # 입금시 입금 내역에 저장
      deposit_count = len(self.deposit_list)+1
      self.deposit_list.append((deposit_count,'입금', money, self.balance))

      # 이자 발생 조건 및 발생시 입금 내역에 저장
      if deposit_count % 5 == 0:
        interest = int(self.balance * 0.01)
        self.balance += interest
        self.deposit_list.append((deposit_count+1,'이자', interest, self.balance))
        print("\n*** SC은행 이자 발생 ***")
        print(f"{interest}의 이자가 발생하였습니다. \n현재 {self.depositor}님의 잔고는 {self.balance:,}원 입니다.")
    else:
      print("\n*** SC은행 입금 현황 ***")
      print(f"1원 이상 입금해주세요. \n현재 {self.depositor}님의 잔고는 {self.balance:,}원 입니다.")

  # 출금
  def withdraw(self, money):
    if money > self.balance:
      print("\n*** SC은행 출금 현황 ***")
      print(f"{self.depositor}님이 출금하려는 금액({money:,}원)이 현재 잔고({self.balance:,}원) 보다 많습니다.")
    else:
      print("\n*** SC은행 출금 현황 ***")
      self.balance -= money
      print(f"{money:,}원 출금 하셨습니다.\n현재 {self.depositor}님의 잔고는 {self.balance:,}원 입니다.")

      #출금시 출금 내역에 저장
      withdraw_count = len(self.withdraw_list)+1
      self.withdraw_list.append((withdraw_count, '출금', money, self.balance))

  # 입금 내역 조회
  def deposit_history(self):
    print(f"\n*** {self.depositor}님의 입금 내역 ***")
    print(f"\n{'횟수':<5} {'내역':<7} {'입금':<14} {'잔액':<10}")
    print("-" * 60)
    for count, text, amount, balance in self.deposit_list:
      print(f"{count:<7} {text:<7} 금액: {amount:<10,} 잔액: {balance:<10,}")

  # 출금 내역 조회
  def withdraw_history(self):
    print(f"\n*** {self.depositor}님의 출금 내역 ***")
    print(f"\n{'횟수':<5} {'내역':<7} {'출금':<14} {'잔액':<10}")
    print("-" * 60)
    for count, text, amount, balance in self.withdraw_list:
      print(f"{count:<7} {text:<7} 금액: {amount:<10,} 잔액: {balance:<10,}")

# 은행 통장 개설
account1 = Account("이윤환", 100000)
account2 = Account("그루", 10000000)
account3 = Account("퍼실", 100000000)

# 총 개설된 계좌 수 출력
account_num = account1.get_account_num()
print(account_num)

# 전체 계좌 리스트 출력
Account.account_display()

# 입금 및 출금 진행(5회 시 이자 자동 발생)
account1.deposit(1000)
account1.deposit(1000)
account1.deposit(1000)
account1.deposit(1000)
account1.deposit(1000)
account1.deposit(10000)
account1.withdraw(1000000)
account1.withdraw(100)
account1.withdraw(100)
account1.withdraw(1000)

# 입금 내역 출력
account1.deposit_history()

#출금 내역 출력
account1.withdraw_history()

# 특정 조건(100만원 이상 보유)시 출력
print('\n',account_num)
print("\n*** 잔고 100만원 이상인 명단 ***")
for account in Account.account_list:
  if account.balance >= 1000000:
    print(f"{account}")

```
